### Supplemental Notebook for DataFrames

Consider two files with several columns in common -- relate this to the `accident.csv` files you have for 1975-2020.

```python
import pandas as pd
```

```python
df1 = pd.read_csv("file01.csv") # file 1
```

```python
df2 = pd.read_csv("file02.csv") # file 2
```

```python
df2
```

`concat` will take two or more dataframes and merge them together.

```python
pd.concat([df1, df2], join="inner", ignore_index=True)
```

Notice that the columns that match in BOTH files are in the new dataframe while the columns NOT shared are eliminated.
